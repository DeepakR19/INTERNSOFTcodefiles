```python
  import matplotlib as mpl
  import matplotlib.pyplot as plt
  import numpy as np
  import pandas as pd
```

```python
plt.plot(np.random.normal(size=300), np.random.normal(size=300), 'd')
```

```python
normals = pd.Series(np.random.normal(size=20))
normals.plot()
```

```python
normals.cumsum().plot(grid=True)
```

```python
variables = pd.DataFrame({'normal': np.random.normal(size=10),
                          'gamma': np.random.normal(size=10),
                          'poisson':np.random.normal(size=10)})
print(variables)
variables.cumsum(0).plot()
```

```python
variables.cumsum(0).plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1, figsize=(4,15))
for i,var in enumerate(['normal','gamma','poisson']):
  variables[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

```python
fig = plt.figure()
ax = plt.axes()
s = np.linspace(0, 20, 100)
print(s)
ax.plot(s, np.sin(s))
```

```python
plt.plot(s, np.sin(s - 0), color='red')
plt.plot(s, np.sin(s - 1), color='y')
plt.plot(s, np.sin(s -2), color='0.55')
```

```python
plt.plot(s, np.sin(s - 0), linestyle=':')
plt.plot(s, np.sin(s - 1), linestyle='-')
plt.plot(s, np.sin(s -2), linestyle='-.')
```

```python
plt.plot(s, np.sin(s - 0))
plt.xlim(-1,15)
plt.ylim(-2,2);
```

```python
plt.plot(s, np.cos(s))
plt.title("Cosine curve")
plt.xlabel("x")
plt.ylabel("cos(x)")
```

```python
rng = np.random.RandomState(7)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,
            cmap='Reds')
plt.colorbar()  # show color scale
```

```python
x = np.linspace(0, 10, 100)
dy = 0.1
y = np.sin(x) + dy * np.random.randn(100)

plt.errorbar(x, y, yerr=dy, fmt='.b');
```

```python
plt.style.use('seaborn-white')

data = np.random.randn(10000)
print(data)
```

```python
plt.hist(data)
```

```python
plt.hist(data, bins=60, normed=True, alpha=0.6, histtype='stepfilled');
```

```python
x = np.random.normal(0, 0.8, 1000)
y = np.random.normal(-2, 1, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.5, normed=True, bins=50)

plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
```

```python
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

```python
mean = [5, 2]
cov = [[3, 1], [2, 0.5]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x, y, bins=20, cmap='Reds')
cb = plt.colorbar()
```

```python
plt.hexbin(x, y, gridsize=20, cmap='Reds')
cb = plt.colorbar()
```
